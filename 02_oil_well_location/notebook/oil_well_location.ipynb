# Обучение и проверка модели для каждого региона

```python
X = geo_data_0.drop(['product', 'id'], axis=1)
y = geo_data_0['product']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=12345)

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

pred_0 = pd.Series(y_pred)
test_0 = y_test.reset_index(drop=True)

print('mean_product', y_pred.mean())
print('RMSE:', mean_squared_error(y_test, y_pred) ** 0.5)
```

# Подсчет рисков и прибыли для каждого региона

```python
budget_per_well = BUDGET / WELLS_COUNT
print('Бюджет на одну сквижну:', budget_per_well)

breakeven_volume = budget_per_well / REVENUE_PER_UNIT
print('Безубыточный объем для одной скважины:', breakeven_volume, 'тыс. баррелей')

print('Регион 0:', pred_0.mean(), 'тыс. бараллей')
print('Регион 1:', pred_1.mean(), 'тыс. бараллей')
print('Регион 2:', pred_2.mean(), 'тыс. бараллей')
```

<div class="alert alert-info">

**Выводы:**

- Безубыточный объем для одной скважины 111.1 тыс. бараллей.
- Средний запас в регионах ниже безубыточного.
- Нужно выбрать 200 лучших скважин чтобы получить прибыль.

</div>

```python
def revenue(ans, pred, count=200):
    top_count = pred.argsort()[-count:]
    total_volume = ans.iloc[top_count].sum()
    return total_volume * REVENUE_PER_UNIT - BUDGET
```

```python
print('Максимальная прибыль для Региона 0:', revenue(test_0, pred_0))
print('Максимальная прибыль для Региона 1:', revenue(test_1, pred_1))
print('Максимальная прибыль для Региона 2:', revenue(test_2, pred_2))
```

<div class="alert alert-info">

**Выводы:**

- **Регион 0:** является лидером по прибыли, которая составляет 3.3 млрд. руб.

- **Регион 1:** является хужшим по прибыли, которая составляет 2.4 млрд. руб.

- **Регион 1:** является средним по прибыли, которая составляет 2.7 млрд. руб.

</div>

```python
def bootstrap(y_test, y_pred, n=1000):
    state = np.random.RandomState(12345)
    profits = []

    for i in range(n):
        subsample = y_pred.sample(n=500, replace=True, random_state=state)
        pred_sorted = subsample.sort_values(ascending=False)
        selected = y_test[pred_sorted.index][:200]
        total_volume = selected.sum()
        profit = total_volume * REVENUE_PER_UNIT - BUDGET
        profits.append(profit)

    return pd.Series(profits)
```

```python
profits_0 = bootstrap(test_0, pred_0)

mean_profit = profits_0.mean()
ci_lower = profits_0.quantile(0.025)
ci_upper = profits_0.quantile(0.975)
risk = (profits_0 < 0).mean() * 100

print(f'Средняя прибыль: {mean_profit:,.0f} руб.')
print(f'95% ДИ: [{ci_lower:,.0f}, {ci_upper:,.0f}] руб.')
print(f'Риск убытков: {risk:.2f}%')
```

```python
profits_1 = bootstrap(test_1, pred_1)

mean_profit = profits_1.mean()
ci_lower = profits_1.quantile(0.025)
ci_upper = profits_1.quantile(0.975)
risk = (profits_1 < 0).mean() * 100

print(f'Средняя прибыль: {mean_profit:,.0f} руб.')
print(f'95% ДИ: [{ci_lower:,.0f}, {ci_upper:,.0f}] руб.')
print(f'Риск убытков: {risk:.2f}%')
```

```python
profits_2 = bootstrap(test_2, pred_2)

mean_profit = profits_2.mean()
ci_lower = profits_2.quantile(0.025)
ci_upper = profits_2.quantile(0.975)
risk = (profits_2 < 0).mean() * 100

print(f'Средняя прибыль: {mean_profit:,.0f} руб.')
print(f'95% ДИ: [{ci_lower:,.0f}, {ci_upper:,.0f}] руб.')
print(f'Риск убытков: {risk:.2f}%')
```
